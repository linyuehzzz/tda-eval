Optimize detailed histogram

```python
# %cd "/content/gdrive/My Drive/Colab Notebooks/census_privacy"
import pandas as pd
import numpy as np
from gurobipy import Model, GRB, QuadExpr, quicksum
import torch
import csv

# define all the input data for the model
filename_cons = 'franklin/microdata/franklin_cons.csv'
cons = pd.read_csv(filename_cons)

# set up parameters
n1, n2, n3, n4, n5 = 8, 2, 2, 63, 2
N = n1 * n2 * n3 * n4 * n5         # number of attribute combinations: HHGQ (8) ∗ VOTINGAGE (2) ∗ HISPANIC (2) ∗ RACE (63) ∗ SEX (2)

with open('franklin/microdata/franklin_hist_obj.csv', 'w', newline='') as f1:
    wr1 = csv.writer(f1)
    wr1.writerow(["GEOID10", "obj"])

    with open('franklin/microdata/franklin_hist.csv', 'w', newline='') as f2:
        wr2 = csv.writer(f2)
        # set up column names
        col_names = ["GEOID10"]
        hist_names = np.empty((n1, n2, n3, n4, n5), dtype="U10")
        for k1 in range(n1):
            for k2 in range(n2):
                for k3 in range(n3):
                    for k4 in range(n4):
                        for k5 in range(n5):
                            hist_names[k1, k2, k3, k4, k5] = str(k1).zfill(2) + str(k2).zfill(2) + str(k3).zfill(2) + str(k4).zfill(2) + str(k5).zfill(2)
        col_names.extend(hist_names.flatten())
        wr2.writerow(col_names)

        for idx, row in cons.iterrows():
            GEOID10 = row["GEOID10"]

            A = torch.tensor(range(N))
            A = A.reshape([n1, n2, n3, n4, n5])

            # initialize model
            m = Model('od')
            m.Params.LogToConsole = 0

            # add objective function
            obj = QuadExpr()

            # add variables and constraints
            h = {}      ## detailed histogram (decision vairable)
            for i in range(N):
                h[i] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, ub=row["P0010001"], name="h_%d"%(i))
            m.update()

            ## P5: HISPANIC OR LATINO ORIGIN BY RACE
            q1 = cons.loc[:, cons.columns.str.startswith("P005")].to_numpy()
            res1, col_idx = {}, 0
            for x in range(n3):  # hispanic
                hist_idx_two_or_more = []
                for y in range(n4):  # race   
                    if y >= 0 and y <= 5:
                        hist_idx = torch.flatten(A[:, :, x, y, :]).tolist()
                        res1[col_idx] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, name="res1_%d"%(col_idx))
                        obj += res1[col_idx] * res1[col_idx]
                        m.addConstr(res1[col_idx] == q1[idx, col_idx] - quicksum(h[i] for i in hist_idx))
                        m.update()
                        col_idx += 1
                    else:
                        hist_idx = torch.flatten(A[:, :, x, y, :]).tolist()
                        hist_idx_two_or_more.extend(hist_idx)
                res1[col_idx] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, name="res1_%d"%(col_idx))
                obj += res1[col_idx] * res1[col_idx]
                m.addConstr(res1[col_idx] == q1[idx, col_idx] - quicksum(h[i] for i in hist_idx_two_or_more))
                m.update()
                col_idx += 1

            ## P8: RACE
            q2 = cons.loc[:, cons.columns.str.startswith("P008")].to_numpy()
            res2, col_idx = {}, 0   
            for x in range(n4):  # race
                hist_idx = torch.flatten(A[:, :, :, x, :]).tolist()
                res2[col_idx] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, name="res2_%d"%(col_idx))
                obj += res2[col_idx] * res2[col_idx]
                m.addConstr(res2[col_idx] == q2[idx, col_idx] - quicksum(h[i] for i in hist_idx))
                m.update()
                col_idx += 1

            ## P9: HISPANIC OR LATINO, AND NOT HISPANIC OR LATINO BY RACE
            q3 = cons.loc[:, cons.columns.str.startswith("P00900")].to_numpy()
            res3, col_idx = {}, 0    
            for x in range(n4):  # race
                hist_idx = torch.flatten(A[:, :, 0, x, :]).tolist()
                res3[col_idx] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, name="res3_%d"%(col_idx))
                obj += res3[col_idx] * res3[col_idx]
                m.addConstr(res3[col_idx] == q3[idx, col_idx] - quicksum(h[i] for i in hist_idx))
                m.update()
                col_idx += 1

            ## P12: SEX BY AGE BY RACE
            q4 = cons.loc[:, cons.columns.str.startswith("P012")].to_numpy()
            res4, col_idx, hist_idx_two_or_more = {}, 0, []    
            for x in range(n4):  # race
                for y in range(n5):  # sex
                    for z in range(n2):  # voting age
                        if x >= 0 and x <= 5:
                            hist_idx = torch.flatten(A[:, z, :, x, y]).tolist()
                            res4[col_idx] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, name="res4_%d"%(col_idx))
                            obj += res4[col_idx] * res4[col_idx]
                            m.addConstr(res4[col_idx] == q4[idx, col_idx] - quicksum(h[i] for i in hist_idx))
                            m.update()
                            col_idx += 1
                        else:
                            hist_idx = torch.flatten(A[:, z, :, x, y]).tolist()
                            hist_idx_two_or_more.extend(hist_idx)
            res4[col_idx] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, name="res4_%d"%(col_idx))
            obj += res4[col_idx] * res4[col_idx]
            m.addConstr(res4[col_idx] == q4[idx, col_idx] - quicksum(h[i] for i in hist_idx_two_or_more))
            m.update()

            ## P43: GROUP QUARTERS POPULATION BY SEX BY AGE BY GROUP QUARTERS TYPE
            q5 = cons.loc[:, cons.columns.str.startswith("P043")].to_numpy()
            res5, col_idx = {}, 0     
            for x in range(n5):  # sex
                for y in range(n2):  # voting age
                    for z in range(n1):  # hhgq
                        hist_idx = torch.flatten(A[z, y, :, :, x]).tolist()
                        res5[col_idx] = m.addVar(obj=0, vtype=GRB.INTEGER, lb=0, name="res5_%d"%(col_idx))
                        obj += res5[col_idx] * res5[col_idx]
                        m.addConstr(res5[col_idx] == q5[idx, col_idx] - quicksum(h[i] for i in hist_idx))
                        m.update()
                        col_idx += 1            

            m.setObjective(obj, GRB.MINIMIZE)
            m.optimize()

            # write histogram values
            hist_values = [GEOID10]
            var_values = [var.X for var in m.getVars() if 'h' == str(var.VarName[0])]
            hist_values.extend(var_values)
            wr2.writerow(hist_values)

            # write objective values
            obj = m.getObjective().getValue()
            wr1.writerow([GEOID10, obj])
            print(idx)
```

```python
import pandas as pd

filename_ppmf = 'franklin/nhgis/ppmf_20210608_franklin.csv'
ppmf = pd.read_csv(filename_ppmf)
ppmf['TABBLKST'] = ppmf['TABBLKST'].apply('{:0>2}'.format)
ppmf['TABBLKCOU'] = ppmf['TABBLKCOU'].apply('{:0>3}'.format)
ppmf['TABTRACTCE'] = ppmf['TABTRACTCE'].apply('{:0>6}'.format)
ppmf['TABBLK'] = ppmf['TABBLK'].apply('{:0>4}'.format)
ppmf['GEOID10'] = ppmf[['TABBLKST', 'TABBLKCOU', 'TABTRACTCE', 'TABBLK']].apply(lambda x: ''.join(x), axis=1)

ppmf.to_csv(filename_ppmf, index=False)
print(ppmf['GQTYPE_PL'].unique(), ppmf['VOTING_AGE'].unique(), ppmf['CENHISP'].unique(), ppmf['CENRACE'].unique())
ppmf
```

Convert to histogram: HHGQ (8) ∗ VOTINGAGE (2) ∗ HISPANIC (2) ∗ RACE (63)

```python
import pandas as pd
import numpy as np
import torch
import csv

filename_hist_new = 'franklin/microdata/franklin_hist.csv'
hist_new = pd.read_csv(filename_hist_new)

filename_ppmf = 'franklin/nhgis/ppmf_20210608_franklin.csv'
ppmf = pd.read_csv(filename_ppmf)

# set up parameters
n1, n2, n3, n4 = 8, 2, 2, 63

with open('franklin/microdata/franklin_hist_ppmf_20210608.csv', 'w', newline='') as f:
    wr = csv.writer(f)
    # set up column names
    col_names = ["GEOID10"]
    hist_names = np.empty((n1, n2, n3, n4), dtype="U10")
    for k1 in range(n1):
        for k2 in range(n2):
            for k3 in range(n3):
                for k4 in range(n4):
                        hist_names[k1, k2, k3, k4] = str(k1).zfill(2) + str(k2).zfill(2) + str(k3).zfill(2) + str(k4).zfill(2)
    col_names.extend(hist_names.flatten())
    wr.writerow(col_names)

    for idx, row in hist_new.iterrows():
        GEOID10 = row['GEOID10']
        ppmf_sub = ppmf[ppmf['GEOID10'] == GEOID10]
        hist_values = []
        for k1 in range(n1):
            for k2 in range(n2):
                for k3 in range(n3):
                    for k4 in range(n4):
                        col_value = len(ppmf_sub[(ppmf_sub['GQTYPE_PL'] == k1) & (ppmf_sub['VOTING_AGE'] == k2 + 1) & (ppmf_sub['CENHISP'] == k3 + 1) & (ppmf_sub['CENRACE'] == k4 + 1)])
                        hist_values.append(col_value)
        hist = [GEOID10]
        hist.extend(hist_values)
        wr.writerow(hist)
        print(idx)
```
